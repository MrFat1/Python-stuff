## UDP Client

```python
import socket
import pickle
import time


ip = "127.0.0.1"
port = 6700
clientid = 1 #Para debug?

#La API esta configurada para mandar 10 preguntas random (multiple o true/false)
for i in range(5):
    
    #Empezamos mandando la petición de pregunta
    msg = "ASK"
    data_pack = (msg, clientid, "Envia pregunta") #Mensaje random para iniciar con algo
    packed = pickle.dumps(data_pack)
    
    #Lo hace todo del tirón
    try:
        server_adress = (ip, port)
        s = socket.socket(family=socket.AF_INET, type=socket.SOCK_DGRAM)
        s.sendto(packed, server_adress)

        #Pide la pregunta
        bytes_rx = s.recvfrom(1024)
        data = pickle.loads(bytes_rx[0])

        print("\nPregunta:\n", data[0])
        print("\nOpciones:")

        opciones = data[1]

        for i in range(len(opciones)):
            print("・", opciones[i])

        respuesta = input("\nElije una de las opciones: ") #Bloqueante

        #Envia la respuesta
        answ_pack = ("ANSWER", clientid, respuesta)
        answ = pickle.dumps(answ_pack)
        s.sendto(answ, server_adress)

        result = s.recvfrom(1024)

        print(pickle.loads(result[0]))
    
    #Server down
    except (ConnectionResetError):
        print("Waiting for server...")
        time.sleep(5)
        print("Timed Out")
        s.close()

end = s.recvfrom(1024)
print(pickle.loads(end[0]))
```
